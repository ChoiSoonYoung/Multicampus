### 14-1-12.txt 텍스트파일을 리스트로 읽어 단어 길이가 5보다 큰 단어만 나열하세요



```python
'''14-1-12.txt 
information
fantastic
activity
such
homeroom
importance
gossip
dictionary
artist
grade
useful
allow
website
rule
communicate'''

with open('./14-1-12.txt', 'r', encoding = 'utf-8') as fd:
    data = fd.read()
    data = data.split('\n')

for i,d in enumerate(data):
    if len(d) <= 5 :
        data.pop(i)
        
print(data)
```

## 연습문제2

### 가위바위보 게임 업그레이드

이전에 만든 가위바위보 게임을 총 게임횟수와 승률을 게임을 다시 실행하도 유지되도록 수정하세요

심화 이름을 입력받아 이름별로 게임횟수와 승률을 저장해주세요


```python
import pickle
import random


#{'김철수': {'전체': 5, '승리': 3, '비김': 1, '승률': 0.6}} 저장 형식
is_first = 0

#처음에는 파일이 없으니까 파일을 만들어줘야함

if is_first == 1:
    user_list = {}

if is_first == 0 :
    with open('./test.pkl', 'rb') as fd :
        data = pickle.load(fd)

#print(data)

#import sys
#sys.exit(1)


game_list = ["가위","바위","보"]
result_dict = {'전체' : 0, '승리' : 0, '비김' : 0}
total = 0


user_name = input()

if user_name not in user_list :
    user_list[user_name] = {'전체': 0,'승리': 0,'비김': 0,'승률': 0,}
#user_name 확인


while(1):
    
    print("가위(1), 바위(2), 보(3) 을 입력해주세요! :")
    user = int(input())
    if user>= 1 and user <=3:
        result_dict['전체'] +=1
        computer = random.randint(1,3)
        print('유저 : {0}, 컴퓨터 : {1}'.format(game_list[user-1],game_list[computer-1]))
        if user == computer :
            result_dict['비김'] +=1
        elif user == 1 and computer == 3 :
            result_dict['승리'] +=1
        elif user == 2 and computer == 1 :
            result_dict['승리'] +=1
        elif user == 3 and computer == 2 :
            result_dict['승리'] +=1
        
    else :
        print('게임종료 (전체:{0}, 승리:{1}, 비김:{2})'.format(result_dict['전체'],result_dict['승리'],result_dict['비김']))
        
        break

user_list[user_name]['전체'] += result_dict['전체']
user_list[user_name]['승리'] += result_dict['승리']
user_list[user_name]['비김'] += result_dict['비김']
user_list[user_name]['승률']= result_dict['승리']/result_dict['전체']

print(user_list)
with open('./test.pkl', 'wb') as f:
    pickle.dump(user_list, f)
    
    
```

다음과 같이 동작하는 프로그램을 개발하세요
프로그램 종료후 다시 접속할때 이전이 입력됬던 값을 파일에 저장해놨다 불러와주세요

심화 수정기능개발
    

```python
import pickle

data_exist = 1

if data_exist == 1:
    with open('./student_score.pkl', 'rb') as fd:
        data = pickle.load(fd)
        student_data = data
if data_exist == 0:
    student_data = []

while (1) :
    
    input_case = int(input('메뉴를 선택해주세요 1 - 입력, 2 - 조회, 3 - 삭제, 4 - 수정, 0 - 종료) :'))

    if input_case == 1 :
        #print('이름 :')
        name = input('이름 :')
        math = int(input('수학 :'))
        science = int(input('과학 :'))
        english = int(input('영어 :'))
        student_data.append({'이름' : name, '수학' : math, '과학' : science, '영어' : english})
    elif input_case == 2 :
        for idx, stu in enumerate(student_data):
            print('[{0}] 이름 : {1}, 수학 : {2}, 과학 : {3}, 영어 : {4}'.format(idx, stu['이름'], stu['수학'], stu['과학'], stu['영어']))
    elif input_case == 3 :
        while(1):
            del_num = int(input('삭제할 번호를 입력해주세요 :'))
            if len(student_data)<del_num :
                print('그런 학생은 없어')
            else:
                student_data.pop(del_num)
                break
    elif input_case == 4 :
        modi_num = int(input('수정할 번호를 입력해주세요 :'))
        print('다음과 같은 데이터가 검색되었습니다. \n[{0}] 이름 : {1}, 수학 : {2}, 과학 : {3}, 영어 : {4}'.format(modi_num, student_data[modi_num]['이름'], student_data[modi_num]['수학'], student_data[modi_num]['과학'], student_data[modi_num]['영어']))
        modi, modi_score = input('수정할 항목과 수정내용을 입력해주세요(ex수학 20) :').split(' ')
        student_data[modi_num][modi] = int(modi_score)
        print('반영되었습니다. \n[{0}] 이름 : {1}, 수학 : {2}, 과학 : {3}, 영어 : {4}'.format(modi_num, student_data[modi_num]['이름'], student_data[modi_num]['수학'], student_data[modi_num]['과학'], student_data[modi_num]['영어']))

        #[0] 이름 : 이민호, 수학 : 40, 과학 : 50, 영어 : 30
    elif input_case == 0 :
        print('종료되었습니다.')
        break
    else:
        print('이상한 숫자 입력하지마 이자식아')

with open('./student_score.pkl', 'wb') as fd:
        pickle.dump(student_data, fd)
```
